```python
import pandas as pd 
import seaborn as sns
```

```python
dataset=pd.read_csv('IRIS.csv')
dataset
```

```python
dataset['species'].unique()
```

```python
dataset['species'].replace({'Iris-setosa':'1', 'Iris-versicolor':'2', 'Iris-virginica':'3'},inplace=True)
dataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(dataset[['sepal_length','sepal_width','petal_length','petal_width']],dataset['species'],test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
lr.predict(X_test)
```

```python
X_test          # Iris-setosa':'1', 'Iris-versicolor':'2', 'Iris-virginica':'3'
```

```python
lr.score(X_test,Y_test)
```

```python
sns.pairplot(dataset[['sepal_length','sepal_width','petal_length','petal_width','species']],hue='species')
```

The above solution is sucessfully predict the species of IRIS Flower by using IRIS dataset, after training the dataset 96% has predicted score.
